```python
def golden_cross(arr1,arr2):
    """
    金叉
    """
    if isinstance(arr2,pd.Series):
        return (arr1.shift(1) < arr2.shift(1)) & (arr1 >= arr2)
    else:
        return (arr1.shift(1) < arr2) & (arr1 >= arr2)

def dead_cross(arr1,arr2):
    """
    死叉
    """
    if isinstance(arr2,pd.Series):
        return (arr1.shift(1) > arr2.shift(1)) & (arr1 <= arr2)
    else:
        return (arr1.shift(1) > arr2) & (arr1 <= arr2)
    
def near_value(arr,stat_value = 0,threshold = 0.05):
    """
    判断前多少个值是否在stat_value附件
    params:
        stat_value: 指定值
        threshold: 阈值
    """
    return (arr - stat_value < threshold) | (arr.shift(1) - stat_value < threshold) | (arr.shift(2) - stat_value < threshold) | (arr.shift(3)- stat_value < threshold) | (arr.shift(4)- stat_value < threshold)
```

```python
from datetime import datetime,timedelta
def get_data(dw,idx):
    if isinstance(idx,str):
        idx_2 = datetime.strptime(idx,'%Y-%m-%d')
    else:
        idx_2 = idx
    return dw[idx_2 - timedelta(days=20):idx_2 + timedelta(days=20)]
```

```python
def plot_k(df):
    df = df[-100:]
    df['over_buy'] = df['slowk'].map(lambda x:90)
    df['over_sell'] = df['slowk'].map(lambda x:10)
    add_plot = [
        mpf.make_addplot(df['slowk'],type='line',color='r',panel=2),
        mpf.make_addplot(df['slowd'],type='line',color='g',panel=2),
        mpf.make_addplot(df['over_buy'],type='line',color='gray',linestyle='--',panel=2),
        mpf.make_addplot(df['over_sell'],type='line',color='gray',linestyle='--',panel=2),
        mpf.make_addplot(df['macdsignal'],type='line',color='r',panel=3),
        mpf.make_addplot(df['macdhist'],type='bar',width=0.7,color='dimgray',panel=3),
        mpf.make_addplot(df['ma60'],type='line',color='r')  # 60日线
    ]
    mpf.plot(df,type='candle',mav=(2,5),addplot=add_plot,volume=True,figscale=1.5,title='KDJ',figratio=(5,5),ylabel='price',ylabel_lower='volume')
```
